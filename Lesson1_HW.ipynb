## 1. Из колоды в 52 карты извлекаются случайным образом 4 карты.  

### a) Найти вероятность того, что все карты – крести.

Крестовых карт всего $^{52}/_4$ = 13.

Вероятность, что первая карта крестовая $^{13}/_{52}$. И последующие соответственно $^{12}/_{51}$,$^{11}/_{50}$,$^{10}/_{49}$.

## $\frac{13}{52}\cdot\frac{12}{51}\cdot\frac{11}{50}\cdot\frac{10}{49}=0.0026$ 

### б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

Здесь используется сумма вероятностей 4-х событий:

## $\frac{4}{52}+\frac{4}{51}+\frac{4}{50}+\frac{4}{49}=0.317$

## 2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

## $C^3_{10}=\frac{10!}{3!7!}=\frac{8\cdot9\cdot10}{1\cdot2\cdot3}=\frac{720}{6}=120$

Число попыток равно 1.

## $P=\frac{1}{120}=0.0083$

## 3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

Вероятность, что первая извлеченная деталь будет окрашена:

## $\frac{9}{15}=\frac{3}{5}$

Вероятность, что 2 извлеченные детали окрашены:

## $\frac{3}{5}\cdot\frac{8}{14}=\frac{12}{35}$

Вероятность, что 3 извлеченные детали окрашены:

## $\frac{12}{35}\cdot\frac{7}{13}=\frac{84}{455}$

## 4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

## $C^2_2=\frac{2!}{2!0!}=1$

## $C^0_{98}=\frac{98!}{0!98!}=1$

## $C^2_{100}=\frac{100!}{2!98!}=\frac{99\cdot100}{1\cdot2}=4950$

## $P=\frac{C^2_2\cdot C^0_{98}}{C^2_{100}}=0.0002$
